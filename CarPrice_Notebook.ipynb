### Sử dụng _LabelEncoder_ để mã hóa dữ liệu thang đo categorical - nominal

### Chuẩn hóa dữ liệu bằng StandardScaler

```python
from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
sdata = s_scaler.fit_transform(data)
col_names = list(data.columns)
sdata = pd.DataFrame(sdata, columns=col_names)
sdata
```

```python
# các thuộc tính định lượng
quantiatives = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']

# các thuộc tính định tính
categorical = ['symboling', 'fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem', 'carbrand', 'carmodel']
```

### Kiểm định tương quan

#### Vẽ biểu đồ heatmap biểu diễn mức độ tương quan giữa các biến định lượng

```python
from matplotlib import pyplot as plt
import seaborn as sns

corr_fields = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']
qtt = sdata.loc[:, corr_fields]
plt.figure(figsize=(15,15))
sns.heatmap(qtt.corr(), annot=True)
plt.show()
```

Biểu đồ heatmap trên cho thấy các biến định lượng: **enginesize, curbweight, horsepower, carwidth, carlength** có tương quan với giá xe **price** (theo thứ tự giảm dần). Tiếp tục thử nghiệm các kiểm định về tương quan:

#### Kiểm định pearsonr đối với các thuộc tính định lượng và giá xe

Đặt ra giả thiết:
- **H0**: Giữa thuộc tính và price không có mối tương quan
- **H1**: Giữa thuộc tính và price có mối tương quan

```python
from scipy.stats import pearsonr

for feature in quantiatives:
    r, pval = pearsonr(sdata[feature], sdata.price)
    if pval < 0.05:
        print(f'{feature} r: {r}- pvalue: {pval}')
```

#### Kết luận: 
Kiểm định _pearsonr_ đối với các thuộc tính **carlength, carwidth, curbweight, enginesize, horsepower, wheelbase, boreratio** có kết quả pvalue < alpha 0.05. Vì thế H1 được chấp nhận => các thuộc tính này _**có tương quan**_ với giá xe. 
Trong số đó có các thuộc tính tương quan ngược với giá xe là **citympg** và **highwaympg** => xe càng tiết kiệm xăng thì giá càng cao

#### Kiểm định spearmanr với các thuộc tính phân loại và giá xe

Đặt ra giả thiết:
- **H0**: Giữa thuộc tính và price không có mối tương quan
- **H1**: Giữa thuộc tính và price có mối tương quan

```python
from scipy.stats import spearmanr

for feature in categorical:
    r, pval = spearmanr(sdata[feature],sdata.price)
    if pval < 0.05:
        print(f'{feature} r: {r}- pvalue: {pval}')
```

#### Kết luận: 
Kiểm định _spearmanr_ đối với các thuộc tính **drivewheel, fuelsystem** có kết quả pvalue < alpha 0.05. Vì thế H1 được chấp nhận => các thuộc tính này _**có tương quan**_ với giá xe

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import pickle

X = data.loc[:, ['carlength', 'carwidth', 'curbweight', 'enginesize', 'horsepower', 'wheelbase', 'boreratio', 'citympg', 'highwaympg', 'drivewheel', 'fuelsystem']]
y = data.price

# dividing data into train and test
x_train, x_test, y_train, y_test = train_test_split(X.values, y.values, test_size=0.2, random_state=7)

# standardzing the data
stds = StandardScaler()
scaler = stds.fit(x_train)
rescaledx = scaler.transform(x_train)
```

```python
# LinearRegression model
LR = LinearRegression()
LR.fit(rescaledx, y_train)

# RandomForestRegressor model
RF = RandomForestRegressor()
RF.fit(rescaledx, y_train)

# DecisionTreeRegressor model
DT = DecisionTreeRegressor()
DT.fit(rescaledx, y_train)
```

```python
LR_y_pred = LR.predict(scaler.transform(x_test))
RF_y_pred = RF.predict(scaler.transform(x_test))
DT_y_pred = DT.predict(scaler.transform(x_test))

print(f'LinearRegression: r2 score: {r2_score(y_test, LR_y_pred)} - MSE: {mean_squared_error(y_test, LR_y_pred)} - RMSE: {np.sqrt(mean_squared_error(y_test, LR_y_pred))}')
print(f'RandomForestRegressor: r2 score: {r2_score(y_test, RF_y_pred)} - MSE: {mean_squared_error(y_test, RF_y_pred)} - RMSE: {np.sqrt(mean_squared_error(y_test, RF_y_pred))}')
print(f'DecisionTreeRegressor: r2 score: {r2_score(y_test, DT_y_pred)} - MSE: {mean_squared_error(y_test, DT_y_pred)} - RMSE: {np.sqrt(mean_squared_error(y_test, DT_y_pred))}')

# saving the trained mode
pickle.dump(RF, open('rf_model.pkl', 'wb'))

# saving StandardScaler
pickle.dump(stds, open('scaler.pkl', 'wb'))
```

```python
def predict(carlength, carwidth, curbweight, enginesize, horsepower, wheelbase, boreratio, citympg, highwaympg, drivewheel, fuelsystem):

    # create data frame
    lists = [carlength, carwidth, curbweight, enginesize, horsepower, wheelbase, boreratio, citympg, highwaympg, drivewheel, fuelsystem]
    
    df = pd.DataFrame(lists).transpose()
    # scaling the data
    scaler.transform(df)
    
    # making predictions using the train model
    prediction = RF.predict(df)
    result = int(prediction)
    return result
```
